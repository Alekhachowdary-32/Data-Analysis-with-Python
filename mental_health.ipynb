```python
#Can we visualize the distribution of different occupations among survey participants?

occupation_counts = health_df['Occupation'].value_counts()
fig = plt.figure(figsize =(16,9))
ax = sns.barplot(x = occupation_counts.index, y = occupation_counts,palette='viridis')
plt.title('Distribution of Occupations')
plt.xlabel("Occupation")
plt.ylabel("Count")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#Mental Health History among the participants
mental_health_counts = health_df['Mental_Health_History'].value_counts()
fig = plt.figure(figsize =(16,9))
ax = sns.barplot(x = mental_health_counts.index, y = mental_health_counts,palette='pastel')
plt.title('Distribution of Mental Health histroy')
plt.xlabel("Mental Health histroy")
plt.ylabel("Count")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#distribution of Work Interest among the participants 
work_interest_counts = health_df['Work_Interest'].value_counts()
fig = plt.figure(figsize =(16,9))
ax = sns.barplot(x = work_interest_counts.index, y = mental_health_counts,palette='husl')
plt.title('Distribution of Work Interest')
plt.xlabel("Work Interest")
plt.ylabel("Count")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#Coping Struggles among the participants
Coping_Struggles_counts = health_df['Coping_Struggles'].value_counts()
fig = plt.figure(figsize =(16,9))
sns.barplot(x = Coping_Struggles_counts.index, y = Coping_Struggles_counts,palette='Blues')
plt.title('Coping Struggles All participants')
plt.xlabel("Coping Struggles")
plt.ylabel("Count")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#Distribution of Quarantine Frustrations based on Mood Swings
colors_quarantine_frustrations_mood_swings = ['lightcoral', 'lightblue', 'lightgreen']
fig = plt.figure(figsize =(16,9))
sns.countplot(x='Mood_Swings', hue='Quarantine_Frustrations', data = health_df, palette=colors_quarantine_frustrations_mood_swings)
plt.xlabel("Mood Swings")
plt.ylabel("Count")
plt.title("Distribution of Quarantine Frustrations based on Mood Swings")
plt.xticks(rotation=45,ha='right')
plt.legend(title='Quarantine Frustrations',title_fontsize='12', loc='upper right')
plt.show()
```

```python
#distribution of 'Weight Change' using a pie chart

weight_change_counts = health_df['Weight_Change'].value_counts()
fig = plt.figure(figsize =(8,8))
plt.pie(weight_change_counts, labels=weight_change_counts.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'), startangle=90)
plt.title('distribution of Weight Change')
plt.show()
```

```python
#distribution of 'Quarantine Frustrations' represented in a pie chart

quarantine_frustrations_counts = health_df['Quarantine_Frustrations'].value_counts()
fig = plt.figure(figsize =(8,8))
plt.pie(quarantine_frustrations_counts, labels=quarantine_frustrations_counts.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'), startangle=90)
plt.title('distribution of Quarantine Frustrations')
plt.show()
```

```python
#visualzing between 'Work Interest', 'Mood Swings', and 'Social Weakness
interest_mood_social_counts = health_df.groupby(['Work_Interest', 'Mood_Swings', 'Social_Weakness']).size().unstack(fill_value=0)
interest_mood_social_counts.plot(kind='bar', stacked=True, colormap='viridis', figsize=(12, 8))
plt.xlabel('Work Interest, Mood Swings, Social Weakness')
plt.ylabel('Count')
plt.title('Relationships between Work Interest, Mood Swings, and Social Weakness')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Variable', title_fontsize='12', loc='upper right')

plt.show()
```
